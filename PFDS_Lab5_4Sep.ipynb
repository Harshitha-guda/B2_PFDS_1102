Merge Product and Sales Datasets and Compute Aggregate
Statistics Using pandas

Create/import two datasets (Products and Sales)

```python
import numpy as np
import pandas as pd
```

```python
ps=pd.read_csv('/content/sales.csv')
ps.head()
```

```python
sd=pd.read_csv('/content/products.csv')
sd.head()
```

Merge them using product_id

```python
merge=pd.merge(ps,sd,on='id')
merge.head()
```

Merge them using product_id.

```python
inner_merge=pd.merge(ps,sd,on='id',how='inner')
inner_merge.head()
```

```python
outer_merge=pd.merge(ps,sd,on='id',how='outer')
outer_merge.head()
```

Compute statistics like total revenue, average price, and
product-wise sales

```python
total_revenue=merge['price'].sum()
print('Total Revenue:',total_revenue)
average_price=merge['price'].mean()
print('Average Price:',average_price)
product_wise_sales=merge.groupby('product_name')['price'].sum()
print('Product-wise Sales:')
print(product_wise_sales)
```

Group by product category and find best-selling category

```python
best_selling_category=merge.groupby('category')['price'].sum().idxmax()
print('Best-selling category:',best_selling_category)
```

Merge Teachers and Courses Datasets and Compute
Aggregate Statistics

Create two datasets (Teachers and Courses)

```python
t=pd.read_csv('/content/teachers.csv')
t.head()
```

```python
c=pd.read_csv('/content/courses.csv')
c.head()
```

Merge on teacher_id

```python
merge=pd.merge(t,c,on='teacher_id')
merge.head()
```

Perform inner and left joins

```python
inner_merge=pd.merge(t,c,on='teacher_id',how='inner')
inner_merge.head()
```

```python
merge_left=pd.merge(t,c,on='teacher_id',how='left')
merge_left.head()
```

Compute average courses per teacher

```python
average_courses_per_teacher=merge['course_id'].mean()
print('Average courses per teacher:',average_courses_per_teacher)
```

Group by department and calculate total workload

```python
group_by_department=merge.groupby('department')['course_id'].count().sum()
print('Total workload per department:')
print(group_by_department)
```
